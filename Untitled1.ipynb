```python
import pandas as pd
from random import shuffle
import re
import numpy as np
```

```python
data = pd.read_csv("hin.txt",sep="\t",header=None)
data.columns = ["english_sentence","hindi_sentence","extra"]
```

```python
data.head(10)
```

```python
data["english_sentence"].apply(lambda x: x.lower())
data["hindi_sentence"].apply(lambda x: x.lower())
```

```python
data["english_sentence"].apply(lambda x: re.sub("''",'',x))
data["hindi_sentence"].apply(lambda x: re.sub("''",'',x))
```

```python
import string
exclude = set(string.punctuation) # Set of all special characters
# Remove all the special characters
data['english_sentence']=data['english_sentence'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
data['hindi_sentence']=data['hindi_sentence'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
```

```python
# Remove all numbers from text
remove_digits = str.maketrans('', '', string.digits)
data['english_sentence']=data['english_sentence'].apply(lambda x: x.translate(remove_digits))
data['hindi_sentence']=data['hindi_sentence'].apply(lambda x: x.translate(remove_digits))

data['hindi_sentence'] = data['hindi_sentence'].apply(lambda x: re.sub("[२३०८१५७९४६]", "", x))

# Remove extra spaces
data['english_sentence']=data['english_sentence'].apply(lambda x: x.strip())
data['hindi_sentence']=data['hindi_sentence'].apply(lambda x: x.strip())
data['english_sentence']=data['english_sentence'].apply(lambda x: re.sub(" +", " ", x))
data['hindi_sentence']=data['hindi_sentence'].apply(lambda x: re.sub(" +", " ", x))
```

```python
data['hindi_sentence'] = data['hindi_sentence'].apply(lambda x : 'START_ '+ x + ' _END')
```

```python
### Get English and Hindi Vocabulary
all_eng_words=set()
for eng in data['english_sentence']:
    for word in eng.split():
        if word not in all_eng_words:
            all_eng_words.add(word)

all_hindi_words=set()
for hin in data['hindi_sentence']:
    for word in hin.split():
        if word not in all_hindi_words:
            all_hindi_words.add(word)
```

```python
x_vocab = len(all_eng_words)
```

```python
y_vocab = len(all_hindi_words)
```

```python
MAX_ENG_LEN = 0
for i in data["english_sentence"]:
    if len(i.split())>MAX_ENG_LEN:
        MAX_ENG_LEN = len(i.split())
MAX_ENG_LEN
```

```python
MAX_HIN_LEN = 0
for i in data["hindi_sentence"]:
    if len(i.split())>MAX_HIN_LEN:
        MAX_HIN_LEN = len(i.split())
MAX_HIN_LEN
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_tokenizer = Tokenizer(num_words=x_vocab)
x_tokenizer.fit_on_texts(data["english_sentence"].values)
X = x_tokenizer.texts_to_sequences(data["english_sentence"].values)
X = pad_sequences(X, maxlen=MAX_ENG_LEN, padding="post")
X.shape
```

```python
y_tokenizer = Tokenizer(num_words=y_vocab)
y_tokenizer.fit_on_texts(data["hindi_sentence"].values)
Y = y_tokenizer.texts_to_sequences(data["hindi_sentence"].values)
Y = pad_sequences(Y,maxlen=MAX_HIN_LEN, padding="post")
Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(X, Y, train_size=0.1, random_state=42, shuffle=True)
```

```python
from tensorflow.keras.layers import LSTM, Embedding, Dense, TimeDistributed
from tensorflow.keras import Input
from tensorflow.keras.models import Model

#making encoder model
embedded_dim = 100
latent_dim = 300

encoder_input = Input(shape=(MAX_ENG_LEN,))
encoder_emb_layer = Embedding(x_vocab, embedded_dim)
encoder_emb = encoder_emb_layer(encoder_input)
encoder_lstm = LSTM(latent_dim, recurrent_dropout=0.4, dropout=0.4, return_sequences=True, return_state=True)
encoder_output, state_h, state_c = encoder_lstm(encoder_emb)

#decoding layer
decoder_input = Input(shape=(None,))
decoder_emb_layer = Embedding(y_vocab, embedded_dim)
decoder_emb = decoder_emb_layer(decoder_input)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_output, decoder_state_h, decoder_state_c = decoder_lstm(decoder_emb, initial_state=[state_h, state_c])

decoder_dense = TimeDistributed(Dense(y_vocab, activation="softmax"))
decoder_output = decoder_dense(decoder_output)

model = Model(inputs=[encoder_input, decoder_input], outputs=decoder_output)
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="rmsprop", metrics=["accuracy"])
```

```python
model.fit([x_tr,y_tr[:,:-1]],y_tr.reshape(y_tr.shape[0],y_tr.shape[1],1)[:,1:], epochs=500, batch_size=128, validation_data=([x_val,y_val[:,:-1]],y_val.reshape(y_val.shape[0],y_val.shape[1],1)[:,1:]))
```

```python
#making inference model for prediction

#for encoder model
encoder_model = Model(inputs=[encoder_input], outputs=[encoder_output,state_h,state_c])

decoder_internal_state_h = Input(shape=(latent_dim,))
decoder_internal_state_c = Input(shape=(latent_dim,))
decoder_internal_states = [decoder_internal_state_h, decoder_internal_state_c]


dec_emb_2 = decoder_emb_layer(decoder_input)
decoder_output, state_h2, state_c2 = decoder_lstm(dec_emb_2, initial_state=[decoder_internal_state_h, decoder_internal_state_c])
decoder_output = decoder_dense(decoder_output)
#for decoder model
decoder_model = Model(inputs=[decoder_input,decoder_internal_states], outputs=[decoder_output,state_h2,state_c2])
```

```python
reverse_target_word_index=y_tokenizer.index_word
reverse_source_word_index=x_tokenizer.index_word
target_word_index=y_tokenizer.word_index
```

```python
def decode_sequence(input_seq):
    e_out, e_h, e_c = encoder_model.predict(input_seq)
    
    target_seq = np.zeros((1,1))
    
    target_seq[0,0] = target_word_index['start']
    
    stop_condition = False
    decoded_seq = ''
    count = 0
    while not stop_condition:
        
        output_tokens,state_h2,state_c2 = decoder_model.predict([target_seq,e_h,e_c])
        
        sampled_token_index = np.argmax(output_tokens)
        sampled_token = reverse_target_word_index[sampled_token_index]
        
        if sampled_token!='end':
            decoded_seq+=' '+sampled_token
            
        if sampled_token == 'end' or len(decoded_seq.split()) >= MAX_HIN_LEN:
            stop_condition = True
        
        target_seq = np.zeros((1,1))
        target_seq[0,0] = sampled_token_index
        
        e_h, e_c = state_h2, state_c2
    return decoded_seq
```

```python
def seq2text(input_seq):
    newString=''
    for i in input_seq:
        if(i!=0):
            newString=newString+reverse_source_word_index[i]+' '
    return newString
```

```python
def seq2hindi(input_seq):
    newString=''
    for i in input_seq:
        if(i!=0):
            newString=newString+reverse_target_word_index[i]+' '
    return newString
```

```python
for i in range(0,100):
    print('Input English sentence:', seq2text(x_val[i]))
    print("real hindi translation: ",seq2hindi(y_val[i]))
    print("Predicted summary:",decode_sequence(x_val[i].reshape(1,MAX_ENG_LEN)))
    print("\n")
```
